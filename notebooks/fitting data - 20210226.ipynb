# Fit Peaks

## File 2/26 17:21

```python
data = np.loadtxt("../data/02_26_sweeper_fwd_avg_17_21_18/sweeper_fwd_avg_17_21_18.txt")
freqs, counts = data
freqs = 1000 * (freqs - 406)

freqs = freqs[80:]
counts = counts[80:]

plt.plot(freqs, counts)

# w_up_1, g_up_1, gamma_up_1, w_c, k_in, k_tot, A, B
popt, cov = fitter.fit_reflection(freqs, counts, 
                            p0=[631, 5.6, 0.1, 528, 20, 33, 15000, 0], 
                        bounds=([600, 1, 0, 400, 0, 0, 5000, 0], 
                                [700, 20, 5, 700, 100, 100, 20000, 5000]))

print(popt)
plt.plot(freqs, fitter.single_siv_ref(freqs, **popt))
plt.show()
```
