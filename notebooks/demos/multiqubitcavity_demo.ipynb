```python
import numpy as np
import matplotlib.pyplot as plt

from sivqed.models.cavity import Cavity, MultiQubitCavity

plt.rcParams['font.size'] = 18
```

```python
# Cavity params is a dict of values for: w_c, k_in, k_out, k_tot
# cavity_params=None uses the default cavity values

# Qubit params is a list of dicts for: w_up/down, g_up/down, gamma_up/down
# qubit_params=None creates 1 qubit with the default values
# qubit_params=[{}, {"w_up": 0}] creates 1 qubit with default and 1 qubit with custom values
m = MultiQubitCavity(cavity_params=None, 
                     qubit_params=[{"w_up": 0}, {}, {"w_up": 50, "g_up": 2}, {"w_up": 150}, {"w_up": 200}])

w_arr = np.linspace(-50, 250, 1000)

plt.plot(w_arr, m.reflectance(w_arr, 1))
```

```python
# set_qubit_params takes a list of dict of the new params, where the keys are the qubit indices
# i.e. 2 refers to the 3rd qubit as specified during initialization.
m.set_qubit_params({2: {"g_up": 5}})

plt.plot(w_arr, m.reflectance(w_arr, 1))
```

```python
# set_cavity_params takes a dict of new cavity parameter
m.set_cavity_params({"k_out": 5})

plt.plot(w_arr, m.transmittance(w_arr, 1))
```
